```python
import csv
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (5,5))
class file:
    def __init__(self, address):
        self.address = address
    def krfiletolist(self):
        filematrix = [] 
        with open(self.address, encoding = 'utf-8') as file:
            myreader = csv.reader(file)
            for i in myreader:
                filematrix.append(i)
        return filematrix
    def filetolist(self):
        filematrix = [] 
        with open(self.address) as file:
            myreader = csv.reader(file)
            for i in myreader:
                filematrix.append(i)
        return filematrix
filematrix1 = file("기온.csv")
filematrix2 = file("식중독.csv")
filematrix1 = filematrix1.krfiletolist()
filematrix2 = filematrix2.krfiletolist()

평균기온_temp = filematrix1[1][1:51] #평균기온은 1973 to 2022.
평균기온 = []
for i in 평균기온_temp:
    평균기온.append(float(i))
식중독건수 = [] # 2002 - 2019
식중독환자수 = []
연도_2002to2019 = []
평균기온_2002to2019 = []

for i in range(2002, 2020):
    연도_2002to2019.append(i)
for i in range(1, len(filematrix2) - 1):
    식중독건수.append(float(filematrix2[i][1]))
for i in range(1, len(filematrix2) - 1):
    식중독환자수.append(int(filematrix2[i][2]))
i = -4
while len(연도_2002to2019) > len(평균기온_2002to2019):
    평균기온_2002to2019.append(평균기온[i])
    i -= 1
평균기온_2002to2019.reverse()
plt.subplots_adjust(hspace=1)

plt.subplot(3, 1, 1)
plt.plot(연도_2002to2019, 평균기온_2002to2019)
plt.title("평균기온(2002-2019)")
plt.xlabel("연도")
plt.ylabel("°C")

plt.subplot(3, 1, 2)
plt.plot(연도_2002to2019, 식중독건수)
plt.title("식중독건수")
plt.xlabel("연도")

plt.subplot(3, 1, 3)
plt.plot(연도_2002to2019, 식중독환자수)
plt.title("식중독환자수")
plt.xlabel("연도")

plt.show()

print('-------------------------------------------------------------------------------------------------')

filematrix3 = file('심장질환사망률.csv')
filematrix3 = filematrix3.filetolist()
#1996 - 2022 허혈성 심장 질환과 기타 심장 질환따로 리스트 만들기
허혈성심장질환사망률_temp = filematrix3[2][4:31]
기타심장질환사망률_temp = filematrix3[3][4:31]
허혈성심장질환사망률 = []
기타심장질환사망률 = []
for i in 허혈성심장질환사망률_temp:
    허혈성심장질환사망률.append(float(i))
for i in 기타심장질환사망률_temp:
    기타심장질환사망률.append(float(i))
연도_1996to2022 = []
for i in range(1996, 2023):
    연도_1996to2022.append(i)
#연도 카운트 -> 해당 개수만큼의 평균기온에서 발췌
평균기온_1996to2022 = []
i = -1
while len(연도_1996to2022) > len(평균기온_1996to2022):
    평균기온_1996to2022.append(평균기온[i])
    i -= 1
평균기온_1996to2022.reverse()

plt.figure(figsize = (8,8))
plt.subplots_adjust(hspace=0.8)

plt.subplot(3, 1, 1)
plt.plot(연도_1996to2022, 평균기온_1996to2022)
plt.title("평균기온(1996-2022)")
plt.xlabel("연도")
plt.ylabel("°C")

plt.subplot(3, 1, 2)
plt.plot(연도_1996to2022, 허혈성심장질환사망률, label = '허혈성심장질환사망률')
plt.title("허혈성심장질환사망률, 기타심장질환사망률")
plt.xlabel("연도")
plt.ylabel("사망률(십만명당)")

plt.subplot(3, 1, 2)
plt.plot(연도_1996to2022, 기타심장질환사망률, label = '기타심장질환사망률')
plt.xlabel("연도")
plt.ylabel("사망률(십만명당)")

plt.legend()
plt.show()
print('-------------------------------------------------------------------------------------------------')

filematrix4 = file('산불통계.csv')
filematrix4 = filematrix4.filetolist()
산불통계 = []#1997 - 2022
산불통계_temp = filematrix4[1][1:27]
연도_1997to2022 = []
평균기온_1997to2022 = [] 

for i in 산불통계_temp:
    산불통계.append(int(i))
for i in range(1997, 2023):
    연도_1997to2022.append(i)

i = -1
while len(연도_1997to2022) > len(평균기온_1997to2022):
    평균기온_1997to2022.append(평균기온[i])
    i -= 1
평균기온_1997to2022.reverse()

plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 1)
plt.plot(연도_1997to2022, 평균기온_1997to2022)
plt.title("평균기온(1997-2022)")
plt.xlabel("연도")
plt.ylabel("°C")

plt.subplot(2, 1, 2)
plt.plot(연도_1997to2022, 산불통계)
plt.title("산불통계")
plt.xlabel("연도")
plt.ylabel("회")

plt.show()
print('-------------------------------------------------------------------------------------------------')

filematrix5 = file('전국폭염일수가공.csv')
filematrix5 = filematrix5.filetolist()
전국폭염일수가공_1973_2022_temp = []
연도_1973_2022 = []
평균기온_1973_2022 = 평균기온[:]
전국폭염일수가공_1973_2022 = []
for i in range(1973, 2023):
    연도_1973_2022.append(i)
for i in filematrix5:
    전국폭염일수가공_1973_2022_temp.append(i[1])
전국폭염일수가공_1973_2022_temp.remove("폭염일수")
for i in 전국폭염일수가공_1973_2022_temp:
    전국폭염일수가공_1973_2022.append(float(i))
    
plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 1)
plt.plot(연도_1973_2022, 평균기온_1973_2022)
plt.title("평균기온(1973-2022)")
plt.xlabel("연도")
plt.ylabel("°C")

plt.subplot(2, 1, 2)
plt.plot(연도_1973_2022, 전국폭염일수가공_1973_2022)
plt.title("전국폭염일수")
plt.xlabel("연도")
plt.ylabel("일")

plt.show()
print('-------------------------------------------------------------------------------------------------')

filematrix6 = file('안면도_이산화탄소_평균농도_원본.csv')
filematrix6 = filematrix6.filetolist()#4열 연평균 
안면도이산화탄소평균농도_1999_2022 = []
평균기온_1999_2022 = []
연도_1999_2022 = []
for i in range(1999, 2023):
    연도_1999_2022.append(i)
i = -1
while len(평균기온_1999_2022) <= 23:
    평균기온_1999_2022.append(평균기온[i])
    i -= 1
평균기온_1999_2022.reverse()
for i in filematrix6:
    if i[3] != '':
        안면도이산화탄소평균농도_1999_2022.append(i[3])
안면도이산화탄소평균농도_1999_2022.remove('연평균')
a = 0
for i in 안면도이산화탄소평균농도_1999_2022:
    안면도이산화탄소평균농도_1999_2022[a] = float(i)
    a += 1
    
plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 1)
plt.plot(연도_1999_2022, 평균기온_1999_2022)
plt.title("평균기온(1999-2022)")
plt.xlabel("연도")
plt.ylabel("°C")

plt.subplot(2, 1, 2)
plt.plot(연도_1999_2022, 안면도이산화탄소평균농도_1999_2022)
plt.title("안면도이산화탄소평균농도")
plt.xlabel("연도")
plt.ylabel("ppm")
```
